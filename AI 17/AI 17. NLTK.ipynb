## 1. Токенизация

## 1.1 Токенизация по словам (Word Tokenization)

```python
text = """Artificial Intelligence is transforming; the way people work and communicate.
Machine learning algorithms analyze large amounts of data to identify patterns and make predictions.
Natural language processing enables computers to understand and generate human language.
Every sentence can be broken down into smaller units called tokens.
These tokens may represent words, subwords, punctuation marks, or even individual characters.
Modern language models process tokens instead of entire sentences, allowing them to understand and generate text efficiently.
Tokenization is one of the fundamental steps in natural language processing.
"""
```

## 2. Лемматизация (Lemmatization)

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words = ["running", "runs", "ran", "better", "cats", "geese", "mice", "wolves"]
print("Лемматизация без указания части речи (По умолчанию существительное)")
print("-"*25)
for word in words:
    lemma = lemmatizer.lemmatize(word)
    print(f"{word:15}: {lemma}")
```

```python
# pos='v' - глагол, pos='n' - существительное, pos='a' - прилагательное, pos='r' - наречие
print("Лемматизация с указание части речи")
print("-"*25)
print(f"running (verb) -> {lemmatizer.lemmatize('running', pos='v')}")
print(f"runs (verb) -> {lemmatizer.lemmatize('runs', pos='v')}")
print(f"ran (verb) -> {lemmatizer.lemmatize('ran', pos='v')}")

print()

print(f"better (adj) -> {lemmatizer.lemmatize('better', pos='a')}")

print()

print(f"cats (noun) -> {lemmatizer.lemmatize('cats', pos='n')}")
print(f"geese (noun) -> {lemmatizer.lemmatize('geese', pos='n')}")
```

```python
# Автоматическое определение части речи для лучшей лемматизации
from nltk import pos_tag
from nltk.corpus import wordnet

def get_wordnet_pos(treebank_tag):
    # NN, VB, JJ, RB
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

sentence = 'The cats are running faster than dogs were walking'

tokens = word_tokenize(sentence)
pos_tagged = pos_tag(tokens)
print("части речи:")
print(pos_tagged)

print("-"* 30)

for word, pos in pos_tagged:
    wordnet_pos = get_wordnet_pos(pos)
    leamma = lemmatizer.lemmatize(word.lower(), pos=wordnet_pos)
    print(f"{word:12} ({pos:4}) -> {leamma}")
```

## 3. Стемминг (Stemming)

[Porter stemmer algorithm](https://ru.wikipedia.org/wiki/%D0%A1%D1%82%D0%B5%D0%BC%D0%BC%D0%B5%D1%80_%D0%9F%D0%BE%D1%80%D1%82%D0%B5%D1%80%D0%B0)

```python
from nltk.stem import (
    PorterStemmer,
    SnowballStemmer,
    LancasterStemmer
)

porter = PorterStemmer()
snowball = SnowballStemmer("english")
lancaster = LancasterStemmer()

words = ["running", "runner", "easily", "fairly", "programming", "programmer", "developer", "programs"]

print(f"{'Слово': <15} {'Porter':<12} {'Snowball':<12} {'Lancaster':<12}")
print("-"*50)
for word in words:
    p = porter.stem(word)
    s = snowball.stem(word)
    l = lancaster.stem(word)
    print(f"{word:<15} {p:<12} {s:<12} {l:<12}")
```

```python
# Stemming vs Lemmatization
print("-"*50)
print(f"{'Слово': <15} {'Стемминг':<15} {'Лемматизация':<15}")
print("-"*50)

comparison_words = ["studies", "studying", "better", "running", 'wolves']

for comparison_word in comparison_words:
    p = lancaster.stem(comparison_word)
    lemma = lemmatizer.lemmatize(comparison_word, pos=wordnet.ADJ)
    print(f"{comparison_word:<15} {p:<12} {lemma}")
```

```python
print("-"*100)
print(SnowballStemmer.languages)

print("-"*100)
russian_stemmer = SnowballStemmer("russian")

russian_words = [
  "бежать",
  "бегун",
  "легко",
  "справедливо",
  "программирование",
  "программист",
  "разработчик",
  "программы"
]

print("Стемминг русских слов")
for russian_word in russian_words:
    s = russian_stemmer.stem(russian_word)
    print(f"{russian_word:<15} {s:<12}")
```

## 5. Полный пайплайн предоброботкт текста

```python
def preprocess_text(text, use_lemmatization=True):
    text = text.lower()

    tokens = word_tokenize(text)

    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words and token.isalpha()]

    if use_lemmatization:
        lem = WordNetLemmatizer()
        pos_tags = pos_tag(tokens)
        tokens = [lem.lemmatize(t, pos=get_wordnet_pos(pos)) for t, pos in pos_tags]
    else:
        stemmer = PorterStemmer()
        tokens = [stemmer.stem(token) for token in tokens]
    return tokens
```

```python
sample_text = """Artificial intelligence has become an important part of modern life, and many people use it every day without even realizing it. From recommendation systems and search engines to virtual assistants and language translation tools, AI is helping users complete tasks more quickly and efficiently. In education, students can use intelligent software to learn new concepts, practice programming, and improve their writing skills. In healthcare, doctors analyze medical data with the help of machine learning algorithms to identify diseases earlier and make better decisions. Businesses rely on data analysis to understand customer behavior and optimize their services. Although artificial intelligence offers many benefits, it also raises questions about privacy, ethics, and security. Researchers continue to develop new models and algorithms while governments and organizations work on regulations to ensure that these technologies are used responsibly. As technology evolves, it is becoming increasingly important for people to understand the basic principles of artificial intelligence and its impact on society.
"""

print(sample_text)

print('-'*50)
print(preprocess_text(sample_text, use_lemmatization=True))

print('-'*50)
print(preprocess_text(sample_text, use_lemmatization=False))
```

```python
# Частотный анализ слов
from nltk import FreqDist

emma_words_lower = [word.lower() for word in emma if word.isalpha()]

freq_dist = FreqDist(emma_words_lower)

print("Top 15 words in Jane Austen: Emma")
print("-"*50)

for word, count in freq_dist.most_common(15):
    print(f"{word:<15}: {count:<15}")

print("-"*50)
hapaxes = freq_dist.hapaxes()
print(len(hapaxes))
print(hapaxes[:10])
```
