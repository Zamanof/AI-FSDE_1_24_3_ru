$$
y = f(w_1x_1 + w_2x_2 + ... + w_nx_n + b)
$$
- $x$ - признаки (входы)
- $w$ - веса
- $f$ - функция активации
- $b$ - смещение(bias)
- $y$ - предсказанный класс



```python
import numpy as np
from sklearn.linear_model import Perceptron
```

### bitwise and
#### 0 and 0 = 0
#### 0 and 1 = 0
#### 1 and 0 = 0
#### 1 and 1 = 1

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])
```

```python
model = Perceptron()
model.fit(X, y)
```

```python
print(model.predict([[0, 0], [1, 1], [1, 0], [1, 0], [0, 1], [1, 1]]))
```
