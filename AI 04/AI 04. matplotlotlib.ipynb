```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10)
```

```python
y = x*2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.title("First plot")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0, 8)
plt.ylim(0, 16)
plt.savefig("plot.png")
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```

```python
a = np.linspace(0, 10, 11)
a
```

```python
b = a**4
b
```

```python
# fig = plt.figure(dpi=50, figsize=(5, 8))
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
axes2 = fig.add_axes([0.5, 0.5, 0.5, 0.5])
axes2.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[1].plot(a, b)
plt.tight_layout()

```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0, 0].plot(x, y)
axes[1, 0].plot(a, b)
plt.tight_layout()
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, label="y=2x", color="#e00dcb")
axes.plot(x, x**2, label="y=x**2", color="blue")
axes.plot(x, x+2, label="y=x+2", color="red", lw=8)
axes.legend(loc="upper right")
# axes.legend(loc=(0, 0.25))
plt.show()
```
