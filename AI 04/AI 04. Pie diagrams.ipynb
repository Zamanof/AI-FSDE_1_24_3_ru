```python
import matplotlib.pyplot as plt
from fontTools.varLib.mutator import percents
```

```python
labels = ["Sumqayit", "Xirdalan", "Baltika"]
percents = [100, 40, 9]
```

```python
plt.pie(percents, labels=labels)
plt.title("Pie diagram")
plt.show()
```

```python
plt.pie(percents, labels=labels, autopct="%1.2f%%")
plt.title("Pie diagram")
plt.show()
```

```python
explode = (0, 0.1, 0)
plt.pie(percents, labels=labels, explode=explode, autopct="%1.2f%%", startangle=90, shadow=True, colors=["red", "green", "blue"])
plt.title("Pie diagram")
plt.show()
```
