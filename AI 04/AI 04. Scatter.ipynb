```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [23, 66, 78, 1, 99, 7]
y = [69, 96, 0, 11, 19, 71]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x = np.random.random(100)
y = np.random.random(100)
colors = np.random.random(100)
sizes = 500 * np.random.random(100)
plt.scatter(x,y, c=colors, s=sizes, alpha=0.8, cmap="inferno")
plt.show()
```

```python
plt.scatter(x,y, color="red", label="red", marker="x")
plt.scatter(y,x, color="blue", label="blue", marker="*", s=sizes)
plt.legend()
plt.show()
```
