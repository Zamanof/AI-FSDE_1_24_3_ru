```python
import matplotlib.pyplot as plt
```

```python
languages = ["Gyurgyshtan", "Nemes", "Fena", "Flomaster"]
words = [10, 15, 25, 2]
```

```python
plt.bar(languages, words, color="#FF0000")
plt.xlabel("Language")
plt.ylabel("Words")
plt.title("World language-words diagram")
plt.show()
```

```python
plt.barh(languages, words, color=["#0054dd", "red", "blue", "orange"])
plt.xlabel("Language")
plt.ylabel("Words")
plt.title("World language-words diagram")
for i, value in enumerate(words):
    plt.text(i, i+0.4, str(value), ha="left")
plt.show()
```
