- DataFrame -> таблица
- Series -> колонки
- read_csv() -> читать из файла(csv)
- head() -> показать первые 5(по умолчанию) или n строк данных
- tail() -> показать последние 5(по умолчанию) или n строк данных
- describe() -> просмотр данных таблицы (mean, std, min, 25%, 50%(median), 75%, max)
- iloc() -> получение строки по номеру позиции
- loc() -> получение строки по имени индекса
