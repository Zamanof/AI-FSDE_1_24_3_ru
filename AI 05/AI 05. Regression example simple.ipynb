```python
import numpy as np
import matplotlib.pyplot as plt
```

## Функция регрессии
$$ y = f(x_1, x_2,..., x_n) + \epsilon $$

- $f$ - функция зависимости
- $y$ - целевая переменная
- $x_i$ - признаки
- $\epsilon$ - ошибка(шум)

## Математическая форма
$$ y = a + bx + \epsilon$$

- $y$ - целевая переменная
- $x$ - независимая переменная (фактор)
- $a$ - свабодный член (intercept)
- $b$ - коофициент наклона(slope)
- $\epsilon$ - случайная ошибка(разница между реальным и предсказанным значением)


```python
x = np.array([1,2,3,4,5])
y = np.array([52, 57, 62, 70, 74])
```

```python
n = len(x)
```

```python
b = (n*np.sum(x*y) - np.sum(x)*np.sum(y))/(n*np.sum(x**2) - np.sum(x)**2)
a = (np.sum(y) - b* np.sum(x)) / n
```

```python
print(f"Уравнение y = {a:.2f} + {b:.2f} * x")
```

```python
y_pred = a+b*x
y_pred
```

```python
plt.scatter(x, y, color="red", label="Данные")
plt.plot(x, y, color="blue", label="Регрессия")
plt.xlabel("Часы подготовки")
plt.ylabel("Оценка")
plt.title("Простая линейная регрессия")
plt.legend()
plt.show()
```

```python
while True:
    try:
        x_new = float(input("Enter hours"))
        if x_new == 0:
            print("Exit")
            break
        y_new = a + b * x_new
        print(f"Prediction score = {y_new:.2f}")
        plt.scatter(x, y, color="red", label="Данные")
        plt.plot(x, y, color="blue", label="Регрессия")
        plt.scatter(x_new, y_new, color="red", label="Our prediction", zorder=5, s=100)
        plt.xlabel("Часы подготовки")
        plt.ylabel("Оценка")
        plt.title(f"Prediction for {x_new} hours")
        plt.legend()
        plt.show()
    except ValueError:
        print("Enter number please")
```
